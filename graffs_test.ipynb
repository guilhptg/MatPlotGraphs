```python
# Data
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]


# Markers settings
marcadores = np.where(vendas == np.min(vendas), 'x', '')


# Cores
cores = plt.get_cmap('Set2').colors
ciclo_cores = cycler('color', cores) 
plt.rc('axes', prop_cycle=ciclo_cores)


# Figure
plt.figure(figsize=(11,6))
plt.plot(meses, vendas, linestyle='--', color='green', linewidth=2)


# Low Marker
for posicao, marcador in enumerate(marcadores):
    if marcador == 'x':
        cor = 'red'
        tamanho = 80
    else:
        cor = cores[3]
        tamanho = 0
    plt.scatter(meses[posicao], vendas[posicao], marker=marcador, s=tamanho, color=cor)
    # Named points
    plt.annotate(vendas[posicao], color=cor, xy=(meses[posicao], vendas[posicao]), xytext=(-7, 12), textcoords='offset points')


# adicionando um vspan para destacar o intervalo de meses com vendas a partir de julho
plt.axvspan(6, 11, color=cores[0], alpha=0.15)

# adicionando texto para destacar o intervalo de meses com vendas a partir de julho
plt.text(6.5, 80, 'Recuperao de crescimento', fontsize=11, color='green', weight='bold')





# X
plt.xlabel('')

# Y
plt.ylabel('Valor em (Milhes)')
plt.yticks([])

#
plt.tick_params(axis='x', length=0)


# 
plt.box(False)
plt.title('Crescimento das Vendas ao Longo do Ano')

plt.show()
```

```python
# DATA  listas de anos e refeies
from cycler import cycler 

anos = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
refeicoes = [40139, 127020, 168193, 153115, 202102, 232897, 277912, 205350, 233389, 232797, 110023, 160350, 222789]

cores = plt.get_cmap('Set3').colors
ciclo_cores = cycler('color', cores)
plt.rc('axes', prop_cycle=ciclo_cores)

plt.figure()

# Cor anotaes
cor_texto = cores[9]

# Lines
plt.plot(anos, refeicoes, linewidth=3, zorder=1)

# Scatter
for posicao in (-1, 0):
    plt.scatter(anos[posicao], refeicoes[posicao], color=cor_texto, zorder=2)
    plt.annotate(refeicoes[posicao], color=cor_texto, xy=(anos[posicao], refeicoes[posicao]), xytext=(10, 0), textcoords='offset points', weight='bold')


# X
plt.xticks(anos)
plt.axvspan(2020, 2021, alpha=0.5, color=cores[5], zorder=0)

# Y
plt.yticks([])

# Ticks Params
plt.tick_params(axis='x', length=0)

plt.text(2020.05, 60000, 'Pandemia', fontsize=10, weight='bold', color=cor_texto)

plt.title('Refeies servidas por ano', fontsize=16, weight='bold', loc='left', color=cores[9])
plt.box(False)

plt.savefig('imagens/anosxrefeicoes.png', bbox_inches='tight', dpi=100)
plt.show()
```
